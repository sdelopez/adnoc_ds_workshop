## Yearly Gas production

How much gas has been produced yearly globally and in every state of the US?

```python
gas_df['Year']=gas_df['Month'].dt.year
```

```python
gas_df['Year']
```

```python
yearly_gas_df = gas_df.groupby(['Year']).sum()
```

```python
yearly_gas_df
```

```python
valid_years = np.logical_and(yearly_gas_df.index >= 2009, yearly_gas_df.index <= 2017)
```

```python
valid_years
```

```python
valid_yearly_gas_df= yearly_gas_df[valid_years]
```

```python
valid_yearly_gas_df
```

```python
valid_yearly_gas_df.filter(items=['U.S.']).plot(kind="bar")
```

### State production

Let's have a look at the yearly production of some specific states

```python
valid_yearly_gas_df.columns[1:].sort_values()
```

```python
valid_yearly_gas_df.filter(items=['Alaska', 'Utah', 'Texas', 'Ohio']).plot(kind="line")
```

## Comparing with Crude Oil Production

```python
file = "data/U.S._crude_oil_production.csv"
crude_df = pd.read_csv(file, decimal=",")
crude_df.head(50)
```

```python
crude_df['Month'] = pd.to_datetime(crude_df['Month'], format="%Y/%m/%d")
```

```python
yearly_crude_df = crude_df.groupby([crude_df['Month'].dt.year]).sum()
```

```python
yearly_crude_df
```

```python
yearly_crude_df.columns
```

```python
yearly_oil=yearly_crude_df['U.S. Crude Oil ']
```

```python
yearly_gas=valid_yearly_gas_df['U.S.']
```

```python
yearly_gas.columns=['Gas']
```

```python
yearly_oil.columns=['Oil']
```

```python
yearly_merged=pd.concat([yearly_oil, yearly_gas], axis='columns')
```

```python
yearly_merged
```

```python
yearly_merged
```

```python
yearly_merged=yearly_merged.drop(yearly_merged.index[9])
```

```python
yearly_merged
```

```python
yearly_merged=yearly_merged.rename(columns={"U.S. Crude Oil ": "Crude", "U.S.": "Gas"})
```

```python
yearly_merged
```

```python
plot=yearly_merged.filter(items=['Crude', 'Gas']).plot(kind="line")
plot.set_xlabel('Year')
plot.legend(['Crude (kboepd)', 'Gas (mcfmpd)'])
```
